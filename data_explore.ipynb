```python
import csv
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline

# Canny Parameters
pLowThreshold=64
pHighThreshold=200
```

```python
samples = []
with open('mydata/driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        samples.append(line)

```

```python
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def hsv(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)
```

```python


fig,ax=plt.subplots(5,4, figsize=(25,20))

ax=ax.ravel()
imgno = 0
for r in range(5):
    index = random.randint(0,len(samples)-1)
    for c in range (1):
        name = 'mydata/IMG/'+samples[index][c].split('/')[-1]
        image = cv2.imread(name)
        ax[imgno].imshow(image)
        ax[imgno].set_title("Orig")
        imgno += 1
        hsvimg = hsv(image)
        h = hsvimg[:,:,0]
        s = hsvimg[:,:,1]
        v = hsvimg[:,:,2]
        ax[imgno].imshow(h,cmap='gray')
        ax[imgno].set_title("H")
        imgno += 1
        ax[imgno].imshow(s,cmap='gray')
        ax[imgno].set_title("S")
        imgno += 1
        ax[imgno].imshow(v,cmap='gray')
        ax[imgno].set_title("V")
        imgno += 1
        #gray = grayscale(image)
        #ax[imgno].imshow(gray,cmap='gray')
        #imgno += 1

        
plt.subplots_adjust(hspace=0.3)
plt.show()
```
